# 유동인구수

#### 월별 유동인구 수를 날짜 별로 묶어준 뒤, 유동인구 수를 제외한 다른 값을 drop하고 일자별로 합산한 값입니다.

일자 | 유동인구수
---|---
날짜 | 유동인구 총 합

```python
Jan = pd.read_csv('./floating_pop/Floating_Population_2001.csv')
Feb = pd.read_csv('./floating_pop/Floating_Population_2002.csv')
Mar = pd.read_csv('./floating_pop/Floating_Population_2003.csv')
Apr = pd.read_csv('./floating_pop/Floating_Population_2004.csv')
May = pd.read_csv('./floating_pop/Floating_Population_2005.csv')
Jun = pd.read_csv('./floating_pop/Floating_Population_2006.csv')
Jul = pd.read_csv('./floating_pop/Floating_Population_2007.csv')
Aug = pd.read_csv('./floating_pop/Floating_Population_2008.csv')
Sep = pd.read_csv('./floating_pop/Floating_Population_2009.csv')
Oct = pd.read_csv('./floating_pop/Floating_Population_2010.csv')
```

```python
Jan_floating = Jan.groupby(['일자']).sum().drop(['시간(1시간단위)','연령대(10세단위)'],axis=1)
Feb_floating = Feb.groupby(['일자']).sum().drop(['시간(1시간단위)','연령대(10세단위)'],axis=1)
Mar_floating = Mar.groupby(['일자']).sum().drop(['시간(1시간단위)','연령대(10세단위)'],axis=1)
Apr_floating = Apr.groupby(['일자']).sum().drop(['시간(1시간단위)','연령대(10세단위)'],axis=1)
May_floating = May.groupby(['일자']).sum().drop(['시간(1시간단위)','연령대(10세단위)'],axis=1)
Jun_floating = Jun.groupby(['일자']).sum().drop(['시간(1시간단위)','연령대(10세단위)'],axis=1)
Jul_floating = Jul.groupby(['일자']).sum().drop(['시간(1시간단위)','연령대(10세단위)'],axis=1)
Aug_floating = Aug.groupby(['일자']).sum().drop(['시간(1시간단위)','연령대(10세단위)'],axis=1)
Sep_floating = Sep.groupby(['일자']).sum().drop(['시간(1시간단위)','연령대(10세단위)'],axis=1)
Oct_floating = Oct.groupby(['일자']).sum().drop(['시간(1시간단위)','연령대(10세단위)'],axis=1)
```

```python
floating_pop = pd.concat([Jan_floating,Feb_floating,Mar_floating,Apr_floating,May_floating,Jun_floating,Jul_floating,Aug_floating,Sep_floating,Oct_floating], join = 'outer')
floating_pop.reset_index()
```

# 코로나 확진자 수

#### 명확한 코로나 바이러스 확진자 데이터가 나타나지 않은 1월은 모두 0으로 두고, 일별로 서울의 확진자 수를 더한 값입니다.

일자 | 확진
---|---
날짜 | 서울 확진자 합계

```python
covid_Feb = pd.read_csv('./covid_confirmed/Feb.csv',encoding = 'CP949')
covid_Mar = pd.read_csv('./covid_confirmed/March.csv',encoding = 'CP949')
covid_Apr = pd.read_csv('./covid_confirmed/April.csv',encoding = 'CP949')
covid_May = pd.read_csv('./covid_confirmed/May.csv',encoding = 'CP949')
covid_Jun = pd.read_csv('./covid_confirmed/June.csv',encoding = 'CP949')
covid_Jul = pd.read_csv('./covid_confirmed/July.csv',encoding = 'CP949')
covid_Aug = pd.read_csv('./covid_confirmed/August.csv',encoding = 'CP949')
covid_Sep = pd.read_csv('./covid_confirmed/September.csv',encoding = 'CP949')
covid_Oct = pd.read_csv('./covid_confirmed/October.csv',encoding = 'CP949')
covid_Nov = pd.read_csv('./covid_confirmed/November.csv',encoding = 'CP949')
```

```python
covid_Mar = covid_Mar[covid_Mar['구분별(1)'] == '시도별']
covid_Mar = covid_Mar[['시점', '확진']].loc[:31,]
covid_Apr = covid_Apr[covid_Apr['구분별(1)'] == '시도별']
covid_Apr = covid_Apr[['시점', '확진']].loc[:30,]
covid_May = covid_May[covid_May['구분별(1)'] == '시도별']
covid_May = covid_May[['시점', '확진']].loc[:31,]
covid_Jun = covid_Jun[covid_Jun['구분별(1)'] == '시도별']
covid_Jun = covid_Jun[['시점', '확진']].loc[:30,]
covid_Jul = covid_Jul[covid_Jul['구분별(1)'] == '시도별']
covid_Jul = covid_Jul[['시점', '확진']].loc[:31,]
covid_Aug = covid_Aug[covid_Aug['구분별(1)'] == '시도별']
covid_Aug = covid_Aug[['시점', '확진']].loc[:31,]
covid_Sep = covid_Sep[covid_Sep['구분별(1)'] == '시도별']
covid_Sep = covid_Sep[['시점', '확진']].loc[:30,]
covid_Oct = covid_Oct[covid_Oct['구분별(1)'] == '시도별']
covid_Oct = covid_Oct[['시점', '확진']].loc[:31,]

covid_Feb.fillna(value=0)
covid_Feb.rename(columns = {'Unnamed: 0' : '시점'}, inplace = True)
covid_Feb.fillna(0, inplace = True)
covid_Feb = covid_Feb[['시점','서울']]
covid_Feb.rename(columns = {'서울' : '확진'}, inplace = True)
```

```python
covid_confirmed = pd.concat([covid_Feb,covid_Mar,covid_Apr,covid_May,covid_Jun,covid_Jul,covid_Aug,covid_Sep,covid_Oct], join = 'outer')
covid_confirmed = covid_confirmed[['시점','확진']]
covid_confirmed
```

# Statistical Analysis 하기 위한 DataFrame 만들기, 회귀분석

```python
import statsmodels.api as sm
```

```python
X1 = sub_pop.merge(floating_pop,on='일자').reset_index()
X2 = covid_values = covid_confirmed['확진'].values.astype(float)
X3 = rainfall_values = rainfall['강수량(mm)'].values
X4 = temp_values = seoul_temp['평균기온(℃)'].values

X = X1.assign(확진=covid_values,강수량=rainfall_values,기온=temp_values).set_index(['일자'])

X
```

# 산점도 행렬

```python
sns.pairplot(X, diag_kind = 'hist')
plt.show;
```

# 전체 관측치의 회귀분석 결과

#### p-value가 큰 '강수량','기온'변수는 제외하고 다시 분석하기로 했습니다.

```python
mod = sm.OLS.from_formula('Passenger~+1+유동인구수+확진+강수량+기온', data= X)
res = mod.fit()
print(res.summary())
```

# 분산 분석

```python
from statsmodels.stats.anova import anova_lm
anova_lm(res)
```

# '강수량' 변수를 제외한 회귀분석 결과

#### 조정계수가 약간 상승했지만 여전히 '기온'은 높은 p-value를 보여주고 있습니다.

```python
mod2 = sm.OLS.from_formula('Passenger~1+유동인구수+확진+기온', data= X)
res2 = mod2.fit()
print(res2.summary())
```

# '기온' 변수를 제외한 회귀분석 결과

#### 마찬가지로 조정계수가 약간 상승했고, '강수량'도 p-value가 높게 나타납니다.

```python
mod3 = sm.OLS.from_formula('Passenger~1+유동인구수+확진+강수량', data= X)
res3 = mod3.fit()
print(res3.summary())
```

# '강수량'과 '기온' 변수 모두 제외한 회귀분석 결과

#### 결정계수는 약간 감소했지만 조정 결정계수는 증가한 것으로 보아, 두 변수를 제외한 회귀분석 식이 적합하다고 판단하였습니다. 

```python
mod4 = sm.OLS.from_formula('Passenger~1+유동인구수+확진', data= X)
res4 = mod4.fit()
print(res4.summary())
```

# 모형의 선형성

```python
fitted = res.predict(X)
residual = X['Passenger'] - fitted
sns.regplot(fitted, residual, lowess=True, line_kws={'color': 'red'})
plt.plot([fitted.min(), fitted.max()], [0, 0], '--', color='grey')
```

# 정규성 검정

```python
import scipy.stats
sr = scipy.stats.zscore(residual)
(x, y), _ = scipy.stats.probplot(sr)

sns.scatterplot(x, y)
plt.plot([-3, 3], [-3, 3], '--', color='grey')
```

# 잔차의 등분산성

```python
sns.regplot(fitted, np.sqrt(np.abs(sr)), lowess=True, line_kws={'color': 'red'})
```

# 다중공선성 검사
